# Parte 2 - Análise Exploratória - Titanic

```python
import matplotlib.pyplot as plt 
import seaborn as sns

titanic = pd.read_csv('titanic.csv')
```

Um novo conjunto de dados nomeado `titanic` foi carregado. Esses dados contêm informações sobre os passageiros do Titanic, incluindo o valor que pagaram pela passagem e se sobreviveram ou não. Vamos investigar se existe uma associação entre a tarifa que um passageiro pagou (`Fare`) e se ele sobreviveu ou não (`Survived`, que é igual a `0` se o passageiro morreu e `1` se sobreviveu): 

- Calcule a diferença na tarifa média paga pelos que sobreviveram e pelos que morreram. Qual grupo pagou uma tarifa média mais alta?

- Calcule a diferença na tarifa mediana para aqueles que sobreviveram e aqueles que morreram.

- Crie um histograma da coluna idade para visualizar a distribuição das idades dos passageiros.

- Crie gráficos boxplot lado a lado de tarifas por sobrevivência. Agora que você pode ver a disseminação dos dados, as diferenças média/mediana parecem relativamente pequenas ou grandes?

- Calcule a matriz de correlação. Existe correlações entre as variáveis?

- Crie gráficos de disperção para melhor visualizar a correlação entre variáveis

- Use suas habilidades e apresente outras análises que ajudaram a entender melhor os dados 

Quem pagou a maior tarifa foi o grupo dos 'Sobreviventes'

```python
mediaSobreviventes = titanic[titanic['survived'] == 1]['fare'].mean()
mediaNaoSobreviventes = titanic[titanic['survived'] == 0]['fare'].mean()

print(f"A média da tarifa paga pelos sobreviventes foi de {mediaSobreviventes:.2f}")
print(f"A média da tarifa paga pelos não sobreviventes foi de {mediaNaoSobreviventes:.2f}")
print(f"A diferença na média de tarifa foi de {(mediaSobreviventes - mediaNaoSobreviventes):.2f}")
```

```python
medianaSobreviventes = titanic[titanic['survived'] == 1]['fare'].median()
medianaNaoSobreviventes = titanic[titanic['survived'] == 0]['fare'].median()

print(f"A mediana da tarifa paga pelos sobreviventes foi de {medianaSobreviventes:.2f}")
print(f"A mediana da tarifa paga pelos não sobreviventes foi de {medianaNaoSobreviventes:.2f}")
print(f"A diferença na mediana de tarifa foi de {(medianaSobreviventes - medianaNaoSobreviventes):.2f}")
```

```python
sns.histplot(data=titanic, x='age', bins=20)
plt.title('Distribuição de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

Como as caixas se sobrepõem consideravelmente e as medianas são próximas, as diferenças podem ser relativamente pequenas. Mas notamos muitos valores discrepantes (outliers) em ambos.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='survived', y='fare')
plt.xticks([0, 1], ['Não Sobreviveu', 'Sobreviveu'])
plt.title('Boxplot de Tarifas por Sobrevivência')
plt.xlabel('Sobrevivência')
plt.ylabel('Tarifa')
plt.show()
```

Nao existe uma forte correlacao entre as variaveis do dataset, podemos notar uma leve correlacao entre 'fare' e 'pclass'.

```python
corr = titanic.corr()
corr.style.background_gradient(cmap='coolwarm').format(precision=2)
```

```python
sns.pairplot(titanic, hue='survived')
plt.title('Gráficos de Dispersão')
plt.show()
```

### Analisar a classe das cabines e como isso afeta as chances de sobrevivência:

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic, x='pclass', hue='survived')
plt.title('Sobrevivência por Classe de Cabine')
plt.xlabel('Classe de Cabine')
plt.legend(['Não Sobreviveu', 'Sobreviveu'])
plt.xticks([0, 1, 2], ['1', '2', '3'])
plt.show()
```

### Verificar se a idade ou a classe das cabines afetam a tarifa paga

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=titanic, x='age', y='fare', hue='pclass')
plt.title('Idade vs. Tarifa por Classe de Cabine')
plt.xlabel('Idade')
plt.ylabel('Tarifa')
plt.legend(title='Classe', labels=['1ª Classe', '2ª Classe', '3ª Classe'])
plt.show()
```
